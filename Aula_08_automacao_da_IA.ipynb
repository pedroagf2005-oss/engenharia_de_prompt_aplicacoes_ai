#AUTOMAÇÃO DE ARQUIVOS MISSÃO 1

```python
import os
import shutil

# Diretório base
base_dir = "arquivos_bagunçados"
os.makedirs(base_dir, exist_ok=True)

# Lista de arquivos fictícios com diferentes extensões
arquivos = [
    "documento1.pdf",
    "imagem1.jpg",
    "musica1.mp3",
    "planilha1.xlsx",
    "texto1.txt",
    "apresentacao1.pptx",
    "codigo1.py",
    "compactado1.zip",
    "video1.mp4",
    "imagem2.png"
]

# Criar os arquivos dentro do diretório base
for arquivo in arquivos:
    caminho = os.path.join(base_dir, arquivo)
    with open(caminho, "w") as f:
        f.write(f"Conteúdo fictício de {arquivo}")

print("Arquivos criados com sucesso!")

# Organizar arquivos em pastas por extensão
for arquivo in arquivos:
    extensao = arquivo.split(".")[-1]
    pasta_destino = os.path.join(base_dir, extensao.upper())  # pasta com nome da extensão
    os.makedirs(pasta_destino, exist_ok=True)

    origem = os.path.join(base_dir, arquivo)
    destino = os.path.join(pasta_destino, arquivo)

    shutil.move(origem, destino)

print("Arquivos organizados por extensão!")
```

#CONSULTA SIMPLES E APIs

```python
import requests

def consultar_cep(cep: str) -> dict:
    """
    Consulta o CEP na API ViaCEP e retorna um dicionário com os dados do endereço
    ou uma mensagem de erro.

    Args:
        cep (str): O número do CEP a ser consultado.

    Returns:
        dict: Um dicionário contendo as informações do endereço (logradouro,
              bairro, localidade, uf, cep) se a consulta for bem-sucedida,
              ou um dicionário com uma chave 'error' e a mensagem de erro
              correspondente em caso de falha.
    """
    url = f"https://viacep.com.br/ws/{cep}/json/"

    try:
        # Adiciona um timeout para a requisição para evitar que ela fique travada
        response = requests.get(url, timeout=5)
        # Levanta um HTTPError para respostas de status de erro (4xx ou 5xx)
        response.raise_for_status()

        dados = response.json()

        if "erro" not in dados:
            return {
                "logradouro": dados.get('logradouro', 'N/A'),
                "bairro": dados.get('bairro', 'N/A'),
                "cidade": dados.get('localidade', 'N/A'),
                "estado": dados.get('uf', 'N/A'),
                "cep": dados.get('cep', 'N/A')
            }
        else:
            return {"error": "CEP não encontrado ou formato inválido."}

    except requests.exceptions.HTTPError as http_err:
        return {"error": f"Erro HTTP: {http_err} - O servidor retornou um erro."}
    except requests.exceptions.ConnectionError as conn_err:
        return {"error": f"Erro de conexão: {conn_err} - Verifique sua conexão com a internet."}
    except requests.exceptions.Timeout as timeout_err:
        return {"error": f"Tempo limite excedido: {timeout_err} - A requisição demorou muito para responder."}
    except requests.exceptions.RequestException as req_err:
        return {"error": f"Erro na requisição: {req_err} - Ocorreu um problema geral com a sua requisição."}
    except ValueError:
        return {"error": "Erro: Resposta da API não é um JSON válido."}
    except Exception as e:
        return {"error": f"Ocorreu um erro inesperado: {e}"}

# --- Exemplo de uso interativo da função refatorada ---
print("\n--- Consulta de CEP Interativa ---")
cep_digitado = input("Por favor, digite o CEP que deseja consultar (apenas números): ")

# Remove qualquer caractere não numérico, caso o usuário digite com hífen
cep_formatado = ''.join(filter(str.isdigit, cep_digitado))

# Validação básica para garantir que o CEP tem 8 dígitos
if len(cep_formatado) != 8:
    print("❌ Erro: O CEP deve conter 8 dígitos numéricos.")
else:
    resultado_interativo = consultar_cep(cep_formatado)
    if "error" not in resultado_interativo:
        print(f"📍 Endereço para o CEP {cep_formatado}:\n"
              f"Logradouro: {resultado_interativo['logradouro']}\n"
              f"Bairro: {resultado_interativo['bairro']}\n"
              f"Cidade: {resultado_interativo['cidade']}\n"
              f"Estado: {resultado_interativo['estado']}\n"
              f"CEP: {resultado_interativo['cep']}")
    else:
        print(f"❌ {resultado_interativo['error']}")
```

#MISSÃO 3 SISTEMA DE NOTIFICAÇÕES

```python
import time
import random

def verificar_condicao(valor: float, limite: float) -> bool:
    """
    Verifica se o valor atual ultrapassa um limite crítico definido.

    Args:
        valor (float): O valor atual a ser verificado.
        limite (float): O limite crítico a ser comparado.

    Returns:
        bool: True se o valor for maior ou igual ao limite, False caso contrário.
    """
    return valor >= limite

def disparar_alerta(valor: float, limite: float) -> None:
    """
    Exibe um alerta crítico no console e simula o envio de uma notificação.

    Args:
        valor (float): O valor atual que disparou o alerta.
        limite (float): O limite que foi ultrapassado.
    """
    mensagem = (
        f"\n🚨 ALERTA CRÍTICO 🚨\n"
        f"Valor atual: {valor:.2f}\n"
        f"Limite permitido: {limite:.2f}\n"
        f"Ação necessária imediatamente!\n"
    )
    print(mensagem)
    # Simulação de envio de e-mail ou outra notificação
    print(f"[Simulação] Enviando e-mail para admin@empresa.com...\n")

def monitorar(limite: float, intervalo: int, duracao_total: int) -> None:
    """
    Monitora continuamente um valor simulado e dispara alertas se o limite for atingido.

    Args:
        limite (float): O limite crítico para o monitoramento.
        intervalo (int): O tempo em segundos entre cada verificação.
        duracao_total (int): A duração total do monitoramento em segundos.
    """
    print(f"Iniciando monitoramento. Limite: {limite:.2f}, Intervalo: {intervalo}s, Duração: {duracao_total}s")
    tempo_inicial = time.time()

    # Loop de monitoramento
    while (time.time() - tempo_inicial) < duracao_total:
        # Simula a coleta de um novo valor (pode vir de um sensor, API, etc.)
        valor_atual = random.uniform(20.0, 100.0) # Valores aleatórios para simulação
        print(f"Monitorando... Valor atual: {valor_atual:.2f}")

        if verificar_condicao(valor_atual, limite):
            disparar_alerta(valor_atual, limite)

        time.sleep(intervalo) # Espera o intervalo definido antes da próxima verificação

    print("Monitoramento encerrado.")

# --- Exemplo de Uso do Sistema de Notificações ---
if __name__ == "__main__":
    LIMITE_CRITICO = 75.0  # Define o limite a partir do qual o alerta será disparado
    INTERVALO_VERIFICACAO = 2 # Verifica a cada 2 segundos
    DURACAO_SIMULACAO = 20 # Simula o monitoramento por 20 segundos

    # Inicia o monitoramento com os parâmetros definidos
    monitorar(LIMITE_CRITICO, INTERVALO_VERIFICACAO, DURACAO_SIMULACAO)
    print("\nFim da simulação de monitoramento.")
```
