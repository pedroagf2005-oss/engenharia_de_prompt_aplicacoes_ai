#Exercício 2: número primo

```python
#prompt original
def eh_primo(n):
    for i in range(2,n):
        if n% i == 0:
            return False
        return True
#prompt corrigido
import math
def eh_primo(n):
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
```

#Exercício 3: Inversão de string

```python
#prompt orginal
def inverte_texto(texto):
    invertido = ''
    for letra in texto:
        invertido = letra + invertido
    return invertido

print(f'nome invertido: {inverte_texto("Pedro")}')

#prompt completo
def  inverte_texto(texto):
    return texto[::-1]
```
